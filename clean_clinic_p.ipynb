```python
import pandas as pd
import numpy as np
import gender_guesser.detector as guess
```

```python
df = pd.read_csv(r"c:/Users/MARIA JOSE/Downloads/m_clinic.csv")
```

```python
df['gender'] = df['gender'].replace(['u', ' '], np.nan)
```

```python
# Guess the gender by the first name :)

g = guess.Detector()

def predict_gender(name):
    first_name = str(name).split()[0].capitalize()
    result = g.get_gender(first_name)

    if 'female' in result: 
        return 'Female'
    if 'male' in result: 
        return 'Male'
    return 'Unknown'
df['gender'] = df['patient_name'].apply(predict_gender)
```
